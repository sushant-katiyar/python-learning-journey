Matplotlib is powerful but verbose. Seaborn wraps around it and gives you two things â€” better looking charts by default, and statistical chart types that Matplotlib doesn't have built in like heatmaps, box plots, and pair plots.
One line of Seaborn often replaces 10 lines of Matplotlib.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Titanic.csv")

# Seaborn has built in themes â€” run this once at the top
sns.set_theme(style="darkgrid")   # options: darkgrid, whitegrid, dark, white, ticks
```

set_theme() makes all your charts look clean automatically. Always call it at the top.

ðŸ§  TOPIC 2 â€” Countplot

Like a bar chart but specifically for counting categories â€” Seaborn does the counting for you:

```python
# How many survived vs died
sns.countplot(x="Survived", data=df, palette="Set3")
plt.title("Survival Count")
plt.show()

# Count by gender, split by survival â€” one line!
sns.countplot(x="Sex", hue="Survived", data=df, palette="Set2")
plt.title("Survival by Gender")
plt.show()
```

hue="Survived" splits each bar by survival status â€” blue for died, orange for survived. This would take 15 lines in Matplotlib. Here it's one. That's Seaborn's power

ðŸ§  TOPIC 3 â€” Histogram with Distribution Curve

```python
sns.histplot(df["Age"], bins=20, kde=True, color="steelblue")
plt.title("Age Distribution")
plt.show()
```

kde=True adds a smooth curve on top of the histogram showing the distribution shape. KDE stands for Kernel Density Estimate â€” it's a smoothed version of the histogram. You'll see this term in ML too.

ðŸ§  TOPIC 4 â€” Boxplot

This is a chart type you haven't seen yet. A boxplot shows the distribution of a numeric column â€” minimum, maximum, median, and outliers all in one chart:

```python
sns.boxplot(x="Pclass", y="Age", data=df, palette="Set3")
plt.title("Age Distribution per Passenger Class")
plt.show()
```

Reading a boxplot:

* The box covers the middle 50% of data (25th to 75th percentile)
* The line inside the box is the median
* The whiskers extend to min and max within a normal range
* Dots beyond the whiskers are outliers



ðŸ§  TOPIC 5 â€” Heatmap

A heatmap shows a table of numbers as colors â€” the stronger the color, the stronger the value. Most commonly used to visualize correlation matrices:

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,          # shows the numbers inside each cell
    fmt=".2f",           # format to 2 decimal places
    cmap="coolwarm",     # color scheme â€” red for positive, blue for negative
    linewidths=0.5
)
plt.title("Correlation Heatmap")
plt.show()
```

annot=True is important â€” without it you just see colors with no numbers. With it you see both. Always use it.

The heatmap will immediately show you which columns correlate with Survived â€” dark red means strong positive correlation, dark blue means strong negative.

ðŸ§  TOPIC 6 â€” Pairplot

This is the most powerful EDA chart. It plots every numeric column against every other numeric column in a grid â€” scatter plots for pairs, histograms on the diagonal:

```python
sns.pairplot(df[["Age", "Fare", "Pclass", "Survived"]], hue="Survived", palette="Set2")
plt.show()
```

hue="Survived" colors dots by survival â€” so you can immediately see if any combination of features separates survivors from non-survivors. This is literally what ML models learn to do automatically.

Pairplot takes a few seconds to render â€” it's generating many charts at once

ðŸ§  TOPIC 7 â€” Violinplot

A violinplot is a boxplot but with the actual distribution shape shown on the sides:

```python
sns.violinplot(x="Survived", y="Age", data=df, palette="muted")
plt.title("Age Distribution by Survival")
plt.show()
```

Wider sections mean more passengers at that age. Thinner sections mean fewer. It combines the statistical summary of a boxplot with the shape information of a histogram.

BOXPLOT â€” Proper Explanation

* Let me use a concrete example. Say you have the ages of 10 passengers:
* 14, 18, 22, 25, 28, 30, 35, 40, 45, 72
* A boxplot summarizes this in 5 numbers:
* Step 1 â€” Sort the data (already done above)
* Step 2 â€” Find the median â€” middle value = 29 (between 28 and 30)
* Step 3 â€” Find Q1 and Q3:

* Q1 (25th percentile) = median of the lower half = 22
* Q3 (75th percentile) = median of the upper half = 40

* Step 4 â€” The box is drawn from Q1 to Q3 â€” so from 22 to 40. This box contains the middle 50% of your data. The line inside the box is the median (29).
* Step 5 â€” The whiskers extend from the box to the furthest point within 1.5 Ã— (Q3 - Q1):

* Q3 - Q1 = 40 - 22 = 18
* Upper whisker goes up to 40 + 27 = 67 maximum
* So 72 is beyond the whisker â€” it becomes a dot â€” an outlier
* 1.5 Ã— 18 = 27

Why it matters in ML

When you saw the Fare boxplot by Pclass â€” first class had many dots way above the whisker. Those are passengers who paid extremely high fares. If you feed those outliers into an ML model without handling them, the model gets confused because it sees one person paying 500 when everyone else pays 30.
Boxplot lets you visually spot this problem before it ruins your model.
