* The **old try/except block** at the bottom. See the problem? Once a function **hits a return**, it exits immediately. Everything after that **else: return** is dead code — it will never execute. Python won't even throw an error, it'll just silently ignore it.

>Correct code

```python
def calculator(x, y, op):
    if op == "+":
        return x + y
    elif op == "-":
        return x - y
    elif op == "*":
        return x * y
    elif op == "/":
        if y == 0:
            return "Cannot divide by zero"
        return x / y
    else:
        return "Invalid operator"

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

print(calculator(num1, num2, op))
```

In Python, you must define a function before you call it. Your version worked because the inputs run first and the function is defined before print(calculator(...)) — but the clean habit is: define functions at the top, run logic at the bottom.
