>Dictionaries: 
 A dictionary stores data as key-value pairs — think of it like a real dictionary where a word (key) has a meaning (value).

```python
student = {
    "name": "Tony",
    "age": 19,
    "branch": "AIML",
    "cgpa": 9.1
}

# Accessing values
print(student["name"])        # Tony
print(student.get("age"))     # 19 — safer way, won't crash if key missing

# Adding / Updating
student["year"] = 1           # adds new key
student["cgpa"] = 9.3         # updates existing key

# Deleting
del student["age"]

# Looping through
for key, value in student.items():
    print(f"{key}: {value}")

# Check if key exists
if "name" in student:
    print("Name exists")
```

>Functions — the full picture

```python
# Default parameters
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Tony"))           # Hello, Tony!
print(greet("Tony", "Hey"))    # Hey, Tony!

# Multiple return values
def get_stats(marks):
    average = sum(marks) / len(marks)
    highest = max(marks)
    lowest = min(marks)
    return average, highest, lowest   # returns a tuple

avg, high, low = get_stats([85, 90, 78, 92])
print(avg, high, low)

# *args — when you don't know how many arguments
def add_all(*numbers):
    return sum(numbers)

print(add_all(1, 2, 3, 4, 5))   # 15
```

>Task 2 — Core Exercise (20 mins)

```python
def get_grade(cgpa):
    if cgpa >= 9:
        return "Outstanding"
    
    elif cgpa >= 8:
        return "Excellent"

    elif cgpa >= 7:
        return "Good"

    elif cgpa >= 6:
        return "Average"

    else:
        return "Need improvement"
    
students = [
    {"name": "Tony", "cgpa": 9.2},
    {"name": "Pepper", "cgpa": 8.1},
    {"name": "Rhodey", "cgpa": 7.4},
    {"name": "Happy", "cgpa": 6.0},
    {"name": "Jarvis", "cgpa": 5.5}
]

for student in students:
    grade = get_grade(student["cgpa"])
    print(f"{student['name']}: {grade}")
```


>Task 3 — THE PROJECT: Contact Book App (45 mins)

```python
contacts = {}

def add_contact(contacts):
    name = input("Enter your name: ")
    phone_no = int(input("Enter your no: "))
    email = input("Enter your email: ")
    contacts[name]={"phone no":phone_no, "email":email}

def view_contacts(contacts):
    for key,value in contacts.items():
        print(f"{key}:{value}")
    

def search_contact(contacts):
    name = input("Enter name: ")
    for name in contacts:
        print(contacts[name])
    else:
        print("No contact found")

def delete_contact(contacts):
    name = input("Enter name: ")
    del(contacts[name])
                                    

def main():
    while True:
        print("\n===== CONTACT BOOK =====")
        print("1. Add Contact")
        print("2. View All Contacts")
        print("3. Search Contact")
        print("4. Delete Contact")
        print("5. Exit")
        
        choice = input("Choose option: ")
        
        if choice == "1":
            add_contact(contacts)
        elif choice == "2":
            view_contacts(contacts)
        elif choice == "3":
            search_contact(contacts)
        elif choice == "4":
            delete_contact(contacts)
        elif choice == "5":
            break

    
        
main()
```
